# Task 2:

We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.
Data: 
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np

max_temperatures = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temperatures = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.arange(12)
plt.plot(months, max_temperatures, 'ro')
plt.plot(months, min_temperatures, 'bo')
plt.xlabel('Months')
plt.ylabel('Temperatures')
plt.show()
```

### Task:

#### 1. Fitting it to the periodic function

```python
from scipy import optimize
def yearly_temperatures(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temperatures,months,max_temperatures, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temperatures,months,min_temperatures, [-40, 20, 0])
```

#### 2. Plot the fit

```python
days=np.linspace(0,12,num=365)
plt.figure()
plt.plot(months,max_temperatures,'ro')
plt.plot(days,yearly_temperatures(days,*res_max),'r-')
plt.plot(months,min_temperatures, 'bo')
plt.plot(days,yearly_temperatures(days,*res_min),'b-')
plt.xlabel('Months')
plt.ylabel('Temperatures ($^\circ$C)')
plt.show()
```
